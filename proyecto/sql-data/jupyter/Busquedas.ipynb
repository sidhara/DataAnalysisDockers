```python
import os
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

connection=mysql.connector.connect(host='sql', user='sloth',passwd='sloth',auth_plugin='mysql_native_password')

cali=pd.read_sql_query('SELECT * FROM database.cali',connection)

cali

fig_dims = (17, 4)
fig, ax = plt.subplots(figsize=fig_dims)

sns.barplot(x=cali['MES'], y=cali['CHOQUE_CON_LESIONADO'], ax=ax)
plt.title('Diagrama de Barras')
plt.xlabel('Meses')
plt.ylabel('Cantidad de Choques con Lesionado')
plt.show()
```

```python
fig_dims = (17, 4)
fig, ax = plt.subplots(figsize=fig_dims)

sns.scatterplot(x=cali['MES'], y=cali['FALSA_ALARMA'], ax=ax, s=100)
plt.title('Gráfica Meses vs Cantidad de Falsas Alarmas')
plt.xlabel('Meses')
plt.ylabel('Cantidad de Falsas Alarmas')
plt.show()
```

```python
fig_dims = (17, 4)
fig, ax = plt.subplots(figsize=fig_dims)
plt.plot(cali['MES'], cali['SOLO_DANOS'], color = 'r')
plt.title('Gráfica Meses vs Cantidad de de solo daños')
plt.xlabel('Meses')
plt.ylabel('Cantidad de solo daños')
plt.show()
```
